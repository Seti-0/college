```python
import math
import numpy as np

def plot(X, Y = None, y_scale=1):

    if Y is None:
        Y = X
        X = list(range(len(Y)))
    
    # Normalize
    X = np.array(X)
    Y = np.array(Y)
    X = X / np.max(X)
    Y = Y / np.max(Y)
    
    y_lim = [min(Y), max(Y)]
    x_lim = [min(X), max(X)]

    points = [[x_lim[1], 0], [x_lim[0], 0]]

    for x, y in zip(X, Y):
        points.append([x, y])

    for i, (x, y) in enumerate(points):
        points[i] = [x*1000, y*1000*y_scale]

    return Polygon(points)
```

```python
X = list(range(100))
Y = list(map(lambda x: x + math.cos(x), X))
plot(X, Y)
```

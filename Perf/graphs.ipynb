```python
import matplotlib.pyplot as plt
import numpy as np

def shadow(X, f):
    
    Xs = np.linspace(min(X), max(X), 200)
    Ys = f(Xs)
    
    plt.plot(Xs, Ys, "--", color="#eeeeee")
    
shadow(np.linspace(-5,5,3), lambda x: x**3)
```

Considering: $P_i = \rho^iP_0$

With $P_0 = 1$

```python

N = np.array(range(10))
p = lambda n: rho**n

plt.figure()
for rho in [0.25, 0.5, 0.75]:
    
    shadow(N, p)
    plt.plot(N, p(N), ".", label=f"$\\rho={rho}$")
    
plt.title("Probability of being in a given state")
plt.xlabel("$n$")
plt.ylabel("$P(S_n)$")
plt.legend()
```

Same again, but normalized properly.

```python
rho = 0.5
n = 10
P0 = (1 - rho) / (1 - rho**(n + 1))
 
I = np.array(range(n))
P = lambda i: P0*rho**i

plt.figure()
for rho in [0.25, 0.5, 0.75]:
    
    shadow(I, P)
    plt.plot(I, P(I), ".", label=f"$\\rho={rho}$")
    
plt.title(f"Probability of being in a given state. ($n={n}$)")
plt.xlabel("$i$")
plt.ylabel("$P(S_i)$")
plt.legend()
```

Looking instead a the chance of being in state 5, vs. rho

```python
n = 10
i = 5

Rho = np.linspace(0, 0.99, 200)

def Pi(i, rho):
    P0 = (1 - rho) / (1 - rho**(n + 1))
    return P0*rho**i

plt.figure()
for i in [1, 2, 3, 4, 5]:
    plt.plot(Rho, Pi(i, Rho), label=f"$S_{i}$")
    
plt.title(f"Probability of being in state i/10")
plt.xlabel("$\\rho$")
plt.ylabel("$P_5$")
plt.legend()
```

Take n infinite, considering the number of customers in the system (and thus the load on the system)

```python
Rho = np.linspace(0, 0.99, 200)
N = lambda rho: rho / (1 - rho)

plt.figure()
plt.plot(Rho, N(Rho))

plt.title(f"System load")
plt.xlabel("$\\rho$")
plt.gca().set_yticks([])
plt.ylabel("Relative Load")
```
